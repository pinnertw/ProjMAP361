## Types, opérations et tests

Il existe trois types numériques distincts: 
- `int` pour les entiers (précision illimitée), 
- `float` pour les réels (nombres flottants, précision dépend de l'architecture),
- `complex` pour les nombres complexes (constante imaginaire pure: `j`).

Il y a un type `bool` pour les expressions booléenes dont les constantes sont `True` et `False`.

La [documentation](https://docs.python.org/fr/3.6/library/stdtypes.html) fournit toutes les informations détaillées.

_Attention:_ il peut y avoir répétition: cette collection n'est pas un ensemble.

## Fonctions

Attention une fonction peut dépendre et interagir avec un contexte global: ce n'est pas une fonction au sens mathématique. 

Une fonction qui ne dépend que de ses arguments est dite _fonction pure_ ou _sans effet de bord_: écriture recommandée.

**Syntaxe:** 
```python
def nom_fonction(arguments):
      instructions
```
Pour renvoyer une valeur on utilise le mot-clé `return`.

### Arguments

- **positional argument**: un argument qui n'est pas suivi par un signe `=` (n'a pas de valeur par défaut)
- **keyword argument**: argument suivi par un signe `=` et une expression qui donne une valeur par défaut

On indique d'abord les _positional arguments_ puis les _keyword arguments_.

````python
def g(a, b, t=0.5):
    z = a + t * (b - a)       # z est une variable locale
    return z
````

````python
print( g(1, 6) )             # t prend la valeur par defaut
print( g(1, 6, 0.2) )        # a et b sont déterminés par la position
print( g(t=0.2, b=6, a=1) )  # arguments déterminés par leurs noms
print( g(b=6, a=1) )         
print( g(1, t=0.2, b=6) )   
print( g(1, b=6) ) 
````

### `*args` et `**kwargs`

__Opérateurs__ `*` __et__ `**`: permettent d'exploser (unpacking) une liste ou un dictionnaire

````python
l = [1, 6]
g(*l)        # appels possibles: g(*l, t=0.2) ou g(t=0.2, *l)
````

````python
d = {'a':1, 'b':6}
g(**d)
````

__Utilisation de__ `*` __et__ `**` __dans les arguments d'une fonction__: permet de regrouper 
- des arguments non-nommés (dans une liste avec `*`, par convention appelée `*args`)
- des arguments nommés (dans un dictionnaire avec `**`, par conversion `**kwargs`)

Très utilisé dans le module `matplotlib`: à connaitre pour bien comprendre la documentation.

__Exemple d'utilisation de__ `*args`:

````python
def gl(*args, t=0.5):
    if len(args) == 2:
        return args[0] + t*(args[1] - args[0])
    else: 
        print('Il faut 2 arguments')
````

````python
gl(1, 6, t=0.2)        # attention l'appel gl(1, 6, 0.2) est invalide !
````

__Exemple d'utilisation de__ `**kwargs`:

````python
def gd(**kwargs):
    if len(kwargs) == 3:
        return kwargs['a'] + kwargs['t']*(kwargs['b'] - kwargs['a'])
    else: 
        print('Il faut 3 arguments')
````

````python
gd(a=1, b=6, t=0.2)    # appel gd(1, 6, t=0.2) invalide
````

### Modification des arguments

Les arguments sont passés par adresse et peuvent dont être modifiés par une fonction (s'ils sont de type modifiable: pas le cas des tuples, `frozenset`...)

````python
def repeat(*args):
    for it in args:
        it *= 2

a = (1, 2)  # tuple non modifiable
b = [1, 2]  # liste modifiable

repeat(a, b)
print(a)
print(b)
````

## Loi des Grands Nombres 

<div style="background-color:#FFEEBB; padding:10px">
<b>Théorème:</b> (Kolmogorov, 1929) <br />
Soit $(X_n)_{n \ge 1}$ des copies indépendantes d'une même variable aléatoire $X$. 
\begin{align*}
\text{Si } \; \mathbf{E}[|X|] < +\infty \quad &\text{alors} \quad S_n = \frac{X_1 + \cdots + X_n}{n} \to \mathbf{E}[X], \\
    \text{et si } \; \mathbf{E}[|X|] = +\infty \quad &\text{alors} \quad S_n = \frac{X_1 + \cdots + X_n}{n} \text{ diverge}. 
\end{align*}
</div>

````python
X = sps.beta(a=2, b=5)
N = 200
x = X.rvs(size = N)

n = np.arange(1, N+1)
Sn = np.cumsum(x) / n
````

````python
fig, ax = plt.subplots(figsize=(16,6))
ax.scatter(n, x, marker="x", color='lightgrey')
ax.axhline(y=X.mean(), color='red', label="Valeur exacte")
ax.plot(n, Sn, label="Valeur de l'estimateur")
ax.set(xlabel = "Itération $n$", 
       ylabel = "Valeur de l'estimateur")
ax.legend(loc='upper right');
````

**Représentation de 100 trajectoires**

````python
fig, ax = plt.subplots(figsize=(16,6))
for M in range(100):
    x = X.rvs(size = N)
    Sn = np.cumsum(x) / n
    ax.scatter(n, x, marker="x", color='lightgrey', alpha=0.2)
    ax.plot(n, Sn, alpha=0.5)
plt.axhline(y=X.mean(), color='red')
plt.xlabel("Itération $n$")
plt.ylabel("Valeur des estimateurs")
plt.ylim((0.1,0.6));
````

## Théorème Central Limite 

<div style="background-color:#FFEEBB; padding:10px">
<b>Théorème:</b> (Laplace, 1812; Lindeberg, 1920) <br />
Soit $(X_n)_{n \ge 1}$ v.a. indépendantes de même loi d’espérance $\mu$ et de variance finie $\sigma^2$. Alors
$$
    \sqrt{n} \biggl(\frac{X_1 + \cdots + X_n}{n} - \mu\biggr) \xrightarrow{\text{loi}} \mathcal{N}\bigl(0, \sigma^2\bigr).
$$
</div>

````python
X = sps.beta(a=2, b=5)
n_iter, N_sample = 200, int(1e5)
sample = X.rvs(size = (n_iter, N_sample))   # n_iter lignes, N_sample colonnes

Sn = np.mean(sample, axis=0)                # moyenne sur chaque ligne
sig = np.std(sample, axis=0)                # écart-type sur chaque ligne
En = np.sqrt(n_iter) * (Sn - X.mean()) / sig
````

````python
xk = np.linspace(np.min(En), np.max(En), 1000)

plt.subplots(figsize=(14,6))
plt.hist(En, bins=50, density=True, label="Histogramme erreur", histtype='step')
plt.plot(xk, sps.norm().pdf(xk), color="red", label=f"Densité N(0,1)")
plt.title(r"Répartition de l'erreur renormalisée $\sqrt{n} (S_n - \mu) / \sigma$")
plt.legend(loc = 'best');
````

## Mesure du temps d'execution avec `%timeit`

````python
import math
times_list = []
n_iters = 10**np.arange(1, 6)
for n in n_iters:
    t = %timeit -o -n20 sum([1./k for k in range(1, n+1)]) - math.log(n)
    times_list.append(t.average)
````

````python
times_np = []
for n in n_iters:
    t = %timeit -o -n20 np.sum(1. / np.arange(1, n+1)) - np.log(n)
    times_np.append(t.average)
````

## Exemple complet

Soit $(G_k)_{k \ge 1}$ une suite _i.i.d._ de variables aléatoires gaussiennes centrées réduites ($G_k \sim \mathcal{N}(0,1)$) et on pose
$$
    X_k = e^{\sigma (G_k + \mu)},
$$
c'est à dire $X_k \sim \mathcal{L}\mathcal{N}(\mu, \sigma)$ (loi log-normale). 

Pour une suite $(X_k)_{k \ge 1}$ on définit $R$ comme le premier temps de passage des sommes partielles de $(X_k)_{k \ge 1}$ au-dessus d'un seuil $L > 0$ fixé, c'est à dire: 
$$
    R = \inf \Bigl\{r \ge 1, \, \sum_{k = 1}^r X_k \ge L \Bigr\}.
$$

Le but est de constuire un **échantillon** de taille $N$ de même loi que $R$, c'est à dire construire $(R_1, \dots, R_N)$, $N$ réalisations indépendantes de même loi que $R$. 

### Version avec `scipy.stats`: *loop_sps*

````python
%%timeit -o      
# on mesure le temps d'execution de la cellule et on renvoit le temps

mu, sigma = 0.1, 0.5
L = 22
X = sps.lognorm(s = sigma, scale = np.exp(mu))

size_N = 1000
sample = []
for i in range(size_N):
    sum_Xk = X.rvs(size=1)
    r = 1
    while sum_Xk < L: 
        sum_Xk += X.rvs(size=1)
        r += 1
    sample.append(r)
````

````python
loop_sps = _      # on récupère la sortie de la cellule précédente
````

### Version avec `numpy.random`: *loop_npr*

````python
%%timeit -o
# on mesure le temps d'execution de la cellule et on renvoit le temps

mu, sigma = 0.1, 0.5
L = 22

size_N = 1000
sample = []
for i in range(size_N):
    sum_Xk = npr.lognormal(mu, sigma, size=1)
    r = 1
    while sum_Xk < L: 
        sum_Xk += npr.lognormal(mu, sigma, size=1)
        r += 1
    sample.append(r)
````

````python
loop_npr = _
````

### Réorganisation de *loop_npr*

````python
def R_simulation(mu, sigma, L):
    sum_Xk = npr.lognormal(mu, sigma, size=1)
    r = 1
    while sum_Xk < L: 
        sum_Xk += npr.lognormal(mu, sigma, size=1)
        r += 1
    return(r)

params = { "mu": 0.1, "sigma" : 0.5, "L": 22 }
size_N = 1000
loop_npr2 = %timeit -o sample = [R_simulation(**params) for i in range(size_N)]
````

La syntaxte de **liste en compréhension** est plus concise que la boucle précédente (celle sur `i` où on fait un `sample.append(r)`) mais le temps d'execution est le même.

### Version batch avec `numpy.random`: *batch_argmax*

On crée une version `batch` pour éviter les appels successifs de `npr.lognormal`. L'idée est de simuler les $(X_k)_{k \ge 1}$ par paquets de longueur $M \ge 2$ (qu'on note `batch_size`): 
$$
    \bigl(X_k, k \ge 0\bigr) = \bigl(\mathbb{X}^M_j, j \ge 0\bigr) \quad \text{avec} \quad \mathbb{X}^M_j = \bigl(X_{j M}, \dots X_{(j +1) M -1}\bigr)
$$

````python
def R_simulation_batch_argmax(mu, sigma, L, batch_size):
    sum_current, r_current = 0, 0
    while True:
        X_j = npr.lognormal(mu, sigma, size = batch_size)
        X_j = np.append(sum_current, X_j)
        S_j = np.cumsum(X_j)
        R = np.argmax(S_j > L)
        if R != 0: 
            return r_current + R
        else:
            sum_current += S_j[-1]
            r_current += batch_size

params = { "mu": 0.1, "sigma" : 0.5, "L": 22 }
size_N = 1000
batch_size = 18
loop_batch_argmax = %timeit -o sample = [R_simulation_batch_argmax(**params, batch_size=batch_size) for i in range(size_N)]
````

### Version batch modifiée avec searchsorted: *batch_search*

````python
def R_simulation_batch_searchsorted(mu, sigma, L, batch_size):
    sum_j, r_j = 0, 0
    while True:
        X_k = npr.lognormal(mu, sigma, size = batch_size)
        X_k = np.append(sum_j, X_k)
        S_k = np.cumsum(X_k)
        R = np.searchsorted(S_k, L)
        if R != batch_size: 
            return r_j + R
        else:
            sum_j = S_k[-1]
            r_j += batch_size

params = { "mu": 0.1, "sigma" : 0.5, "L": 22 }
size_N = 1000
batch_size = 18
loop_batch_search = %timeit -o sample = [R_simulation_batch_searchsorted(**params, batch_size=batch_size) for i in range(size_N)]
````

### Version sans boucle: *sample_argmax*

````python
def R_sample_batch_argmax(mu, sigma, L, batch_size, size_N):
    sums, r = np.zeros(size_N), 0
    result = []
    while True:
        X_k = npr.lognormal(mu, sigma, size=(batch_size, len(sums)))
        X_k = np.insert(X_k, 0, sums, axis=0)
        S_k = np.cumsum(X_k, axis=0)
        R = np.argmax(S_k > L, axis=0)
        result = np.concatenate((result, r + R[R>0]))
        if np.sum(R == 0) == 0:
            return result
        else:
            sums, r = S_k[-1, R==0], r+batch_size

params = { "mu": 0.1, "sigma" : 0.5, "L": 22 }
size_N = 1000
batch_size = 18
sample_batch_argmax = %timeit -o sample = R_sample_batch_argmax(**params, batch_size=batch_size, size_N=size_N)
````
